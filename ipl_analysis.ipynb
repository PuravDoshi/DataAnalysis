```python
# Which team has most number of wins?
df.head()
df12=df['winner'].value_counts().reset_index()
df12.rename(columns={'winner':'index','count':'wins'},inplace=True)
df12
```

```python
matches_played = df['team1'].value_counts() + df['team2'].value_counts()
```

```python
matches_played
```

```python
df11=matches_played.to_frame().reset_index()
df11
```

```python
df13=df11.merge(df12,on='index',how='left')
df13
```

```python
x=df13['index']
y=df13['count']
z=df13['wins']

width=0.4
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Teams",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.title("Winnings",fontsize=15)

plt.bar(p,y,width,color='red',label='Matches Played')
plt.bar(p1,z,width,color='yellow',label='Matches Won')
plt.xticks(p+width/2,x,rotation=90)

plt.legend()
plt.show()
```
